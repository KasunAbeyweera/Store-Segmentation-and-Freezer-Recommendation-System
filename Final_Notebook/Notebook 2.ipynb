## Model Building

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, MinMaxScaler, QuantileTransformer
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Load data into a pandas DataFrame
df = pd.read_csv('final.csv')
df['Outlet_ID'] = le.fit_transform(df['product_type'])
# Drop the "Outlet_ID" column
# df = df.drop('Outlet_ID', axis=1)
df = df.drop('area(sqft)', axis=1)

# Scale the data
# scaler = MinMaxScaler()
# scaled_data = scaler.fit_transform(df)
scaled_data =df
```

```python
inertia = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
    print("For n_clusters={}, The Inertia is {}".format(k, kmeans.inertia_))

plt.plot(range(2, 11), inertia)
plt.title('Inertia')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.show()
```

```python
# Silhouette coefficient for optimal k value
silhouette = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    label = kmeans.labels_
    silhouette.append(silhouette_score(scaled_data, kmeans.labels_))
    sil_coeff = silhouette_score(scaled_data, label, metric='euclidean')
    print("For n_clusters={}, The Silhouette Coefficient is {}".format(k, sil_coeff))

plt.plot(range(2, 11), silhouette)
plt.title('Silhouette Coefficient')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.show()

```

```python
# Davies-Bouldin index for optimal k value
davies_bouldin = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    label2 = kmeans.labels_
    davies_bouldin.append(davies_bouldin_score(scaled_data, label2))

plt.plot(range(2, 11), davies_bouldin)
plt.title('Davies-Bouldin Index')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.show()
```

```python
# Calinski-Harabasz Index for optimal k value
calinski_harabasz = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    chs = calinski_harabasz_score(scaled_data, kmeans.labels_)
    calinski_harabasz.append(chs)
    print("For n_clusters={}, The Calinski-Harabasz Index is {}".format(k, chs))

plt.plot(range(2, 11), calinski_harabasz)
plt.title('Calinski-Harabasz Index')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.show()
```

```python
import itertools

# List of shops and their volumes and sales
shops = {
    'Cluster 1': {'volume': 654.6255026809652, 'sales': 674141.9302949061},
    'Cluster 2': {'volume': 831.6625595238095, 'sales': 922939.4717261904},
    'Cluster 3': {'volume': 513.53262169967, 'sales': 296813.5457920792}
}

# List of freezers
freezers = [
    {'name': 'ChillMaster', 'model': 'M001', 'capacity': 55, 'cost': 2500},
    {'name': 'FreezeZone', 'model': 'M002', 'capacity': 75, 'cost': 2800},
    {'name': 'CoolTech', 'model': 'M003', 'capacity': 120, 'cost': 3200},
    {'name': 'mini CoolTech', 'model': 'M004', 'capacity': 30, 'cost': 1800},
    {'name': 'IceBlast Pro', 'model': 'M005', 'capacity': 150, 'cost': 3500},
    {'name': 'IceBlast', 'model': 'M006', 'capacity': 65, 'cost': 2700},
    {'name': 'FreezeTech', 'model': 'M007', 'capacity': 100, 'cost': 3000},
    {'name': 'FreezeTech 3.0', 'model': 'M008', 'capacity': 25, 'cost': 1600},
    {'name': 'ChillTech', 'model': 'M009', 'capacity': 80, 'cost': 2800},
    {'name': 'IceTech', 'model': 'M010', 'capacity': 200, 'cost': 4500}
]

# Function to calculate item return on investment
def item_roi(sales, freezer_cost, volume):
    return (sales / freezer_cost * 1.68)

# Function to get all combinations of freezers
def get_freezer_combinations(num_freezers):
    return itertools.combinations(freezers, num_freezers)

# Function to find the optimal combination of freezers for a shop category
def find_optimal_combination(shop_category):
    best_roi = 0
    best_combination = None
    for i in range(1, len(freezers) + 1):
        for combination in get_freezer_combinations(i):
            freezer_capacities = sum([f['capacity'] for f in combination])
            freezer_cost = sum([f['cost'] for f in combination])
            roi = item_roi(shops[shop_category]['sales'], freezer_cost, shops[shop_category]['volume'])
            if roi > best_roi:
                best_roi = roi
                best_combination = combination
    return best_combination

# Find the optimal combination of freezers for each shop category
for shop_category in shops:
    best_combination = find_optimal_combination(shop_category)
    freezer_names = [f['name'] for f in best_combination]
    print(f"Shop Category: {shop_category}")
    print(f"Optimal Freezer Combination: {freezer_names}")
    print()
```
